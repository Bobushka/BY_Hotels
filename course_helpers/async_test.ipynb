```python
import asyncio
import aiohttp
```

```python
async def get_data(i: int, endpoint: str):
    url = f"http://127.0.0.1:8000/{endpoint}/{i}"
    print(f"Start of request: {i}")
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as resp:
            print(f"End of request: {i}")
```

```python
# Запуск программы
await get_data(1, "sync")
```

```python
func = "async"  # or "async"
n = 300  # start from 3 -> 30 -> 300
await asyncio.gather(
    *[get_data(i, func) for i in range(n)]  # распаковка массива чтобы не писать много раз
)
```
